# Import Liabraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

# Load Dataset

```python
df = pd.read_csv(
    r"C:\Users\HP\Downloads\archive (2)\RUvideos.csv",
    encoding='latin1',
    low_memory=False)
df.head()
```

# Data Cleaning

```python
print(df.columns)
```

```python

df = df.dropna()
cols = ['views', 'likes', 'dislikes', 'comment_count']
for col in cols:
    df[col] = pd.to_numeric(df[col], errors='coerce')


df = df[df['views'] > 0]
df.head()
```

# Engagement Rate Calculation

```python
df['engagement_rate'] = (df['likes'] + df['comment_count']) / df['views']
df.head()
```

# Top Performing Videos

```python
top_videos = df.sort_values(by='engagement_rate', ascending=False).head(10)
print(top_videos[['title', 'views', 'engagement_rate']])
```

# Category-Wise Performance

```python
category_perf = df.groupby('category_id')[['views', 'likes', 'comment_count']].mean()
print(category_perf)
```

# Visualisation

# Engagement Rate Distribution

```python
sns.histplot(df['engagement_rate'], bins=50)
plt.title("Engagement Rate Distribution")
plt.show()
```

# Views VS Likes

```python
plt.scatter(df['views'], df['likes'])
plt.xlabel("Views")
plt.ylabel("Likes")
plt.title("Views vs Likes")
plt.show()
```

# Top Categories

```python
category_perf['views'].sort_values(ascending=False).head(10).plot(kind='bar')
plt.title("Top Categories by Views")
plt.show()
```

# Conclusion

Videos with higher likes & comments have better engagement.
Some videos have high views but low engagement вҶ’ weak content quality.
Certain categories consistently perform better.
Engagement rate is a better metric than views alone.
